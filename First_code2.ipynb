```
import matplotlib.pyplot as plt # 간단히 plt로 사용
plt.title('plotting')     # 그래프 제목
plt.plot([10,20,30,40])   # y축의 값
plt.show()
```

```
plt.plot([1,2,3,4],[12,43,25,15]) # x,y축 데이터 모두 사용
plt.show()
```

```
plt.title('legend')
plt.plot([10,20,30,40], label = 'asc')
plt.plot([40,30,20,10], label = 'desc')
plt.legend(loc=5)
```

```
plt.title('linestyle')
plt.plot([10,20,30,40],color='r', linestyle='--', label = 'dashed')
plt.plot([40,30,20,10],color='g', ls=':', label = 'dotted')
plt.legend(loc=8)
plt.show()
```
