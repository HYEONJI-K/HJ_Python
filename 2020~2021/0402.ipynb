```python
import matplotlib.pyplot as plt
from matplotlib import rc, font_manager
import numpy as np
import csv
```

```python
f = open('/gdrive/MyDrive/2021 Data analysis and Machine Learning/0402/seoul.csv', 'r', encoding = 'cp949')
data = csv.reader(f)
header = next(data)
result = []
for row in data:
  if row[-1] != '':
    if row[0].split('-')[1]=='08':  #날짜 데이터 - 기준 split 0 년도 1: 월 2/-1: 일
      result.append(float(row[-1]))
f.close()
print(result)
plt.figure(figsize=(20,3))
plt.rc('font', family='NanumBarunGothic')
plt.title("1907년-2000년까지 서울의 8월달 최고기온 그래프")
plt.xlabel("최고기온")
plt.ylabel("인스턴스의 개수")
plt.plot(result, 'olive')
plt.show()
```

```python
# 서울의 가장 최고기온 찾기
max_temp = -999
max_date = ''

f = open('/gdrive/MyDrive/2021 Data analysis and Machine Learning/0402/seoul.csv', 'r', encoding = 'cp949')
data = csv.reader(f)
header = next(data)

for row in data:
  if row[-1] == '':
    row[-1] = -999
  row[-1] = float(row[-1])

  if max_temp < row[-1]:
    max_temp = row[-1]
    max_date = row[0]  # 날짜 데이터

f.close()
print("기상 관측 이래 서울의 최고기온 검색")
print("날짜 : {}, 최고기온 : {}". format(max_date, max_temp))
```

```python
# 서울의 가장 최고기온 찾기
min_temp = 999
min_date = ''

f = open('/gdrive/MyDrive/2021 Data analysis and Machine Learning/0402/seoul.csv', 'r', encoding = 'cp949')
data = csv.reader(f)
header = next(data)

for row in data:
  if row[3] == '':
    row[3] = 999
  row[3] = float(row[3])

  if min_temp > row[3]:
    min_temp = row[3]
    min_date = row[0]  # 날짜 데이터

f.close()
print("기상 관측 이래 서울의 최저기온 검색")
print("날짜 : {}, 최저기온 : {}". format(min_date, min_temp))
```
