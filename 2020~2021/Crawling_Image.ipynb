웹페이지에서 마우스 오른쪽 버튼 사용 불가한 이미지 가져오기

```python
import requests
from PIL import Image
```

```python
# header 옵션을 이용하여 이미지 주소에 접근한다.

URL = "https://image-comic.pstatic.net/webtoon/728015/215/20210511151411_e773f69621f030e3303a5204ad784a21_IMAG01_2.jpg"
headers = {
    "Referer" : "https://comic.naver.com/webtoon/detail.nhn?titleId=728015&no=215&weekday=sat",
    "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/66.0.3359.139 Safari/537.36",
}
# request 패키지의 header를 활용하면 Python을 통해 접속했음을 숨기고 요청할 수 있다.
response = requests.get(URL, headers=headers)
```

```python
# Image 패키지를 통해 응답으로 받은 이미지를 실제 화면에 출력한다.
# Image 패키지는 응답으로 받은 이미지를 확인하기 위해 활용된다.
with open("test.png", "wb") as f:
    f.write(response.content)
img = Image.open("test.png")
img
```
