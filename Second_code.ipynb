```
import matplotlib.pyplot as plt
import csv
```

```
# 구글 드라이브에 있는 seoul.csv 파일을 읽어 온다.
f = open('/gdrive/My Drive/Colab Notebooks/seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data)
result = []

for row in data:
  if row[-2] != "": # 최고기온 데이터 값이 존재한다면
    result.append(float(row[-2])) # result 리스트에 최고기온 값을 추가
print(result)
f.close()
plt.figure(figsize=(20,5))
plt.title("Seoul Temperature")
plt.plot(result, 'g', linestyle = '--')
plt.show()
```

```
# 구글 드라이브에 있는 seoul.csv 파일을 읽어 온다.
f = open('/gdrive/My Drive/Colab Notebooks/seoul.csv', encoding='cp949')
data = csv.reader(f)
next(data)
result = []

for row in data:
  if row[-1] != '': # 최고기온 데이터 값이 존재한다면
    if row[0].split('-')[1]=='08':
      result.append(float(row[-1])) # result 리스트에 최고기온 값을 추가
print(result)
f.close()
plt.figure(figsize=(20,5))
plt.title("August Temperature")
plt.plot(result, 'lightseagreen', linestyle = '--')
plt.show()
```
