```python
f = open("crawling test.txt", "w", encoding='utf-8')

raw = requests.get("https://www.data.go.kr/tcs/dss/selectDataSetList.do?dType=FILE&keyword=&detailKeyword=%28%EC%9B%90%EC%A3%BC%EC%8B%9C%29&publicDataPk=&recmSe=&detailText=%28%EC%9B%90%EC%A3%BC%EC%8B%9C%29&relatedKeyword=&commaNotInData=&commaAndData=&commaOrData=%EC%9B%90%EC%A3%BC%EC%8B%9C&must_not=&tabId=&dataSetCoreTf=&coreDataNm=&sort=_score&relRadio=&orgFullName=&orgFilter=&org=&orgSearch=&currentPage=1&perPage=40&brm=&instt=&svcType=&kwrdArray=&extsn=&coreDataNmArray=&pblonsipScopeCode=")
html = BeautifulSoup(raw.text,"html.parser")
container = html.select("div.result-list > ul > li")
for i in range(0,40):
  title = container[i].select_one("dl").text
  s = container[i].select_one("div.info-data").text
  f.write(title+","+s)
f.close()
```
