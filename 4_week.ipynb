### 📌 Q1. 시점 컬럼에서 연도와 분기에 대한 파생변수를 생성하고 기술통계를 구해주세요.

- 데이터 컬럼을 단위인 "백만"으로 변경해 주세요.

```python
# 한글폰트 사용을 위해 설치
# 아래 모듈을 설치하고 불러오면 별도의 한글폰트 설정이 필요 없습니다.
!pip install koreanize-matplotlib
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import koreanize_matplotlib

df = pd.read_csv(
    "https://raw.githubusercontent.com/corazzon/boostcourse-ds-510/master/data/online_oversea_sale_202210.csv.zip", 
    encoding="cp949")
df.shape
```

```python
df.info()
```

```python
from glob import glob
# 데이터는 '백만',
# 시점에서 "." 을 기준으로 x[0] 은 '연도'
# 시점에서 "."을 기준으로 x[1], "/" 을 기준으로 x[0] 은 '분기'로 분리
df["백만"] = df["데이터"]
df["연도"] = df["시점"].map(lambda x : int(x.split(".")[0]))
# inf 값이 나옴 -> nan 변경 -> int화
df["분기"] = df["시점"].map(lambda  x : x.split(".")[1].split("/")[0]).replace("-", np.nan).astype(int)
df.head()
```

```python
# 원하는 데이터(백만, 연도, 분기)가 int화 된 것을 확인!
df.info()
```
